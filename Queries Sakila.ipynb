```python
import pandas as pd
q = """SELECT *
FROM city, customer_list
WHERE city_id = 312;"""
pd.read_sql(q, engine)
```

```python
a = """SHOW tables;"""
pd.read_sql(a, engine)
```

```python
c = """SELECT *
FROM category, film
WHERE name = "Comedy";"""
pd.read_sql(c, engine)
```

```python
e = """SELECT DISTINCT *
FROM actor, film
WHERE first_name = 'Johnny' AND last_name = 'Lollobrigida';"""
pd.read_sql(e, engine)
```

```python
g = """SELECT DISTINCT *
FROM actor, film
WHERE title = 'Bingo Talented';"""
pd.read_sql(g, engine)
```

```python
f = """SELECT DISTINCT amount, customer_id
FROM payment
WHERE amount >
    (SELECT 2 * AVG(amount)
    FROM payment);"""
pd.read_sql(f, engine)
```

```python
i = """SELECT customer.first_name, customer.last_name,
    (SELECT COUNT(*)
    FROM payment
    WHERE customer.customer_id = payment.customer_id) AS num_payments
FROM customer
ORDER by num_payments DESC
LIMIT 5;"""
pd.read_sql(i, engine)
```
